```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import textwrap
import json
```

```python
try:
    from googlesearch import search
except ImportError:
    print("No module named 'google' found")
```

```python
df=pd.read_csv('/content/grematvocab - Sheet1.csv')
```

```python
all_vocab=df.values.ravel()
```

```python
len(all_vocab)
```

```python
mapper={}
```

```python
np.where(all_vocab=='entrenched')
```

```python
left=[]
for vocab in all_vocab[640:660]:
  query = '"'+vocab+'" the economist'
  all_links=[]
  for j in search(query, num=10, stop=10):
    if ('gre-vocabulary' not in j) and ( vocab not in j) and ( 'www.economist.com' in j):
      all_links.append( j )
  if len(all_links)==0:
    left.append( vocab )
  
  for URL in all_links:
    try:
      if URL not in mapper.keys():
        page = requests.get(URL)
        soup = BeautifulSoup(page.content, "html.parser")
        js = json.loads(soup.select_one("script[type=application/ld+json]").text)
        if vocab in soup.text:
          blank='. '.join(js['articleBody'].split(vocab)[0].split('.')[-2:]) + ' _______ ' + '. '.join(js['articleBody'].split(vocab)[1].split('.')[:2])
        else:
          blank='. '.join(js['articleBody'].split(vocab.title())[0].split('.')[-2:]) + ' _______ ' + '. '.join(js['articleBody'].split(vocab.title())[1].split('.')[:2])
        blank = textwrap.dedent(blank).strip()
        mapper[URL]=blank
      else:
        blank=mapper[URL]

      #print(textwrap.fill(blank, width=90))
      res=np.concatenate([np.random.choice( all_vocab[all_vocab!=vocab] ,2),np.array([vocab])])
      np.random.shuffle(res)
      print('<div class="card"><div class="front"><p> ',
          textwrap.fill(blank, width=90)+'<ol><li>',res[0],'</li><li>',res[1],'</li><li>',res[2],'</li></ol>'+'</p></div><div class="back"><p>',
        vocab,'<a href=',URL,'>Source</a></dd> </p></div></div>')
      #print(''.join(['#']*100))
      print('\n')
    except:
      try:
        if URL not in mapper.keys():
          page = requests.get(URL)
          soup = BeautifulSoup(page.content, "html.parser")
          if vocab in soup.text:
            blank='. '.join( soup.text.split(vocab)[0].split('.')[-2:] ) + '_______' + '. '.join( soup.text.split(vocab)[1].split('.')[:2] )
          else:
            blank='. '.join( soup.text.split(vocab.title())[0].split('.')[-2:] ) + '_______' + '. '.join( soup.text.split(vocab.title())[1].split('.')[:2] )
          blank = textwrap.dedent(blank).strip()
          mapper[URL]=blank
        else:
          blank=mapper[URL]

        res=np.concatenate([np.random.choice( all_vocab[all_vocab!=vocab] ,2),np.array([vocab])])
        np.random.shuffle(res)
        print('<div class="card"><div class="front"><p> ',
          textwrap.fill(blank, width=90)+'<ol><li>',res[0],'</li><li>',res[1],'</li><li>',res[2],'</li></ol>'+'</p></div><div class="back"><p>',
              vocab,'<a href=',URL,'>Source</a></dd> </p></div></div>')
        #print(''.join(['#']*100))
        print('\n')
      except:
        continue
```
